```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patheffects
```

```python
# based on https://matplotlib.org/stable/gallery/images_contours_and_fields/contours_in_optimization_demo.html#sphx-glr-gallery-images-contours-and-fields-contours-in-optimization-demo-py
fig, ax = plt.subplots(figsize=(6, 6))

nx = 101
ny = 105

# Set up survey vectors
xvec = np.linspace(0.001, 4.0, nx)
yvec = np.linspace(0.001, 4.0, ny)


x1, x2 = np.meshgrid(xvec, yvec)

# Evaluate some stuff to plot
obj = x1**2 + x2**2 - 2*x1 - 2*x2 + 2# x1**2 + x1*x2 - x2**2

cntr = ax.contour(x1, x2, obj, [0.01, 0.1, 0.5, 1, 2, 4, 8, 16],
                  colors='black')
ax.clabel(cntr, fmt="%2.1f", use_clabeltext=True)

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

nx = 101
ny = 101

# Set up survey vectors
xvec = np.linspace(-4.0, 4.0, nx)
yvec = np.linspace(-4.0, 4.0, ny)


x1, x2 = np.meshgrid(xvec, yvec)

# Evaluate some stuff to plot
obj = x1**2 + x1*x2 - x2**2

# cntr = ax.contour(x1, x2, obj, [4.0])
cntr = ax.contour(x1, x2, obj)
ax.clabel(cntr, fmt="%2.1f", use_clabeltext=True)
ax.grid(True)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

a = 1.0
b = 1.05

nx = 101
ny = 101

# Set up survey vectors
xvec = np.linspace(-2.0, 2.0, nx)
yvec = np.linspace(-1.0, 1.0, ny)


x1, x2 = np.meshgrid(xvec, yvec)

# Evaluate some stuff to plot
obj = ((x1-a)**2 + x2**2)*((x1+a)**2 + x2**2)

cntr = ax.contour(x1, x2, obj, [0.9**4, 1.0**4, 1.05**4])

ax.clabel(cntr, fmt="%2.1f", use_clabeltext=True)
ax.grid(True)
```
