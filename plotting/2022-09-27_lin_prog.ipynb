```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

ax.plot([9,0], [0,9]) # x1+x2 <9
ax.plot([0,10], [6,6]) # x2 < 6
ax.plot([7,7], [0,10]) # x1 < 7
ax.plot([3,6], [9,0]) # 3x1 + x2 < 18

#ax.plot([0,10], [0,0], ':k')
#ax.plot([0,0], [0,10], ':k')

ax.plot([0, 10],[7, 3.75], 'y')
plt.text(8, 4.5, '$H$')

fr = plt.Polygon([(0,0), (6,0), (4.5,4.5), (3,6), (0,6)], facecolor='0.9', edgecolor='0.5')
ax.add_patch(fr)

fig.text(0.9 ,0.05,'$x_1$')
fig.text(0.05, 0.9, '$x_2$')

plt.text(-0.5, 7.5, '$x_1 + x_2 \leq 9$')
plt.text(8, 5.5, '$x_2 \leq 6$')
plt.text(5.8, 9.0, '$x_1 \leq 7$')
plt.text(0.9, 8.5, '$3x_1 + x_2 \leq 18$')

plt.text(0.1, 5.6, '$v_4$')
plt.text(2.8, 5.5, '$v_3$')
plt.text(3.9, 4.2, '$v_2$')
plt.text(5.5, 0.2, '$v_1$')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)


plt.savefig("test.png", dpi=600)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

nx = 101
ny = 101

# Set up survey vectors
xvec = np.linspace(0.0, 5.0, nx)
yvec = np.linspace(0.0, 5.0, ny)


x1, x2 = np.meshgrid(xvec, yvec)

# Evaluate some stuff to plot
obj = 3*x1 + 2*x2

cntr = ax.contour(x1, x2, obj, [5, 10, 15, 20, 22.5],
                  colors='black')
ax.clabel(cntr, fmt="%2.1f", use_clabeltext=True)

ax.plot([9,0], [0,9]) # x1+x2 <9
ax.plot([0,10], [6,6]) # x2 < 6
ax.plot([7,7], [0,10]) # x1 < 7
ax.plot([3,6], [9,0]) # 3x1 + x2 < 18

ax.plot([4.5, 4.5],[0, 4.5], ':k')
ax.plot([0, 4.5],[4.5, 4.5], ':k')


fr = plt.Polygon([(0,0), (6,0), (4.5,4.5), (3,6), (0,6)], facecolor='0.9', edgecolor='0.5')
ax.add_patch(fr)

fig.text(0.9 ,0.05,'$x_1$')
fig.text(0.05, 0.9, '$x_2$')

plt.text(-0.5, 7.5, '$x_1 + x_2 \leq 9$')
plt.text(8, 5.5, '$x_2 \leq 6$')
plt.text(5.8, 9.0, '$x_1 \leq 7$')
plt.text(0.9, 8.5, '$3x_1 + x_2 \leq 18$')

plt.text(0.1, 5.6, '$v_4$')
plt.text(2.8, 5.5, '$v_3$')
plt.text(3.9, 4.2, '$v_2$')
plt.text(5.5, 0.2, '$v_1$')

plt.savefig("test.png", dpi=600)
```
