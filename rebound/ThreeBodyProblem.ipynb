```python
import rebound
import matplotlib.pyplot as plt
import numpy as np
```

```python
sim = rebound.Simulation()
sim.add(m=1)
sim.add(m=1e-1, x = 1.0, vx = 0.8, y = 0.8, vy = 0.)
sim.add(m=1e-1, x = 1.3, vx = 0.2, y = -1., vy = 0.)
```

```python
Nout = 10000
times = np.linspace(0,20.*2.*np.pi,Nout)
x = np.zeros((sim.N, Nout))
y = np.zeros((sim.N, Nout))

ps = sim.particles

for ti, t in enumerate(times):
    sim.integrate(t)
    for i, p in enumerate(ps):
        x[i][ti] = p.x
        y[i][ti] = p.y      
```

```python
%matplotlib inline
for i in range(0, sim.N):
    plt.plot(x[i], y[i])

plt.grid(True)
```
