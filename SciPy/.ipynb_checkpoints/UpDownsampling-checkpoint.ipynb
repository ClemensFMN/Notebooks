```python
import numpy as np
import scipy.signal as sig
import scipy.fft as fft
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
def plt_fft(u):
    ufft = fft.fftshift(fft.fft(u))
    ufft_abs =  10*np.log10(ufft.real**2 + ufft.imag**2)
    plt.plot(fft.fftshift(fft.fftfreq(len(u))), ufft_abs)
```

```python
def plot_response(w, h, title):
    "Utility function to plot response functions"
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(w/2/np.pi, 20*np.log10(np.abs(h)))
    ax.set_ylim(-60, 5)
    ax.set_xlim(0, 0.5)
    ax.grid(True)
    ax.set_xlabel('Normalized Frequency')
    ax.set_ylabel('Gain (dB)')
    ax.set_title(title)
```

```python
cutoff = 0.22 # Desired cutoff frequency
trans_width = 0.05  # Width of transition from pass band to stop band
numtaps = 51      # Size of the FIR filter.

taps = sig.remez(numtaps, [0, cutoff, cutoff + trans_width, 0.5], [1, 0], fs=1.0)
w, h = sig.freqz(taps, [1], worN=2000)

plot_response(w, h, "Lowpass")
```

```python
N = 500
t = np.linspace(0, 1, N)
u = sig.sawtooth(2 * np.pi * 20 * t)
u = sig.lfilter(taps, [1], u)
plt.plot(t,u)
```

```python
plt_fft(u)
```

```python
v = u[::2]
plt_fft(v)
```

```python
uhat = np.zeros(N)
uhat[::2] = v
# plt.plot(uhat)
# uhat
plt_fft(uhat)
```

```python
uhat_filter = 2*sig.lfilter(taps,[1],uhat)
plt.plot(t,uhat_filter,'r-', t,u,'-g')
```

```python
plt_fft(uhat_filter)
```
