```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = 1.
W = 0.1
```

```python
x = np.linspace(0, H, 100)
```

```python
plt.plot(x, x + np.sqrt((H-x)**2 + W**2))
plt.grid(True)
```

```python
plt.savefig('crossing_rectangle_1.png')
```

```python
alpha = np.linspace(0, 2, 100)
```

```python
r = (1 + alpha)/np.sqrt(1 + alpha**2)
```

```python
plt.plot(alpha, r)
plt.grid(True)
```
