```python
from sympy import *
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 7]
```

```python
a1=2
a2=4
N = 50
```

```python
x = Symbol('x')
```

```python
f=1/((1-x**a1)*(1-x**a2))
```

```python
sf=series(f, n=N)
```

```python
coeffs = [sf.coeff(x**i) for i in range(1,N)]
```

```python
plt.plot(coeffs, '-rx')
plt.grid(True)
```
