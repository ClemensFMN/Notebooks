```python
from sympy import *
init_session()

from sympy.plotting import plot
%matplotlib inline
```

We have a model for a stock over 3 years: It rises for two years (by x % every year) and then falls for one year (by x %). Starting with an initial model of 1, the value after 3 years is therefore f.

```python
f = (1+x)**2 * (1-x)
```

```python
plot(f,xlim=(-2,1), ylim=(0,2))
```

```python
d = diff(f,x)
d
```

```python
solve(d,x)
```
