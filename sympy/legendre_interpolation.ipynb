# Interpolation using Legendre Polynomials

define the polynomais

```python
L0 = legendre_poly(0,x)
L1 = legendre_poly(1,x)
L2 = legendre_poly(2,x)
L3 = legendre_poly(3,x)
L0, L1, L2, L3
```

Scale so that they become orth**normal** polynomials

```python
phi0 = L0 / sqrt(integrate(L0**2, (x,-1,1)))
phi1 = L1 / sqrt(integrate(L1**2, (x,-1,1)))
phi2 = L2 / sqrt(integrate(L2**2, (x,-1,1)))
phi3 = L3 / sqrt(integrate(L3**2, (x,-1,1)))
```

Expand the function $f(x) = \sin(x)$

I don't know how we can assign this to a variable...

```python
c0 = integrate(sin(x) * phi0, (x,-1,1))
c1 = integrate(sin(x) * phi1, (x,-1,1))
c2 = integrate(sin(x) * phi2, (x,-1,1))
c3 = integrate(sin(x) * phi3, (x,-1,1))
```

```python
p = c0*phi0 + c1*phi1 + c2*phi2 + c3*phi3
p
```

Compare function and approximation

```python
plot(sin(x), p, (x,-2,2))
```
