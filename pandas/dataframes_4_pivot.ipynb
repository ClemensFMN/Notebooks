```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 3,
        "B": ["A", "B", "C"] * 4,
        "D": range(12),
    }
)
df #.head()
```

```python
pd.pivot_table(df, values="D", index=["A", "B"], aggfunc='sum')
```

```python
pd.pivot_table(df, values="D", index="A", columns="B", aggfunc='sum')
```
