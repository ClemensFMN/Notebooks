```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.DataFrame(
    {
        "col1": ["one", "one", "two", "three"] * 3,
        "col2": ["A", "B", "C"] * 4,
        "col3": range(12),
    }
)
df #.head()
```

```python
# aggfunc is applied to the column in values
# the index is the group by
pd.pivot_table(df, values="col3", index=["col1"], aggfunc='sum')
```

```python
# if we have several cols in index, it groups by all of them
pd.pivot_table(df, values="col3", index=["col1", "col2"], aggfunc='sum')
```

```python
# order of cols in index matters!
pd.pivot_table(df, values="col3", index=["col2", "col1"], aggfunc='sum')
```

```python
# we can use a different aggregation function
pd.pivot_table(df, values="col3", index=["col1"], aggfunc='max')
```

```python
# and use more than one aggregation function
pd.pivot_table(df, values="col3", index=["col1"], aggfunc=['sum', 'max'])
```

```python
# instead of the hierarchies created by index=["col2", "col1"] above, we can make it 2-d
pd.pivot_table(df, values="col3", index="col1", columns="col2", aggfunc='sum')
```
