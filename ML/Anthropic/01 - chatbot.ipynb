## Model Selection

I took the modelId from a bedrock tutorial & it seems that this is the only one working... In addition, I have to use the `us-east-1` region... Other regions do not work. I have no idea why => TODO

```python
model_id = "us.anthropic.claude-haiku-4-5-20251001-v1:0"
```

```python
client = boto3.client("bedrock-runtime", region_name="us-east-1")
```

```python
# helper functions - https://anthropic.skilljar.com/claude-in-amazon-bedrock/276722

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": [
            {"text": text}
        ]
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant", 
        "content": [
            {"text": text}
        ]
    }
    messages.append(assistant_message)

def chat(messages):
    response = client.converse(
        modelId=model_id,
        messages=messages
    )
    return response["output"]["message"]["content"][0]["text"]
```

## Simple Example of a conversation

```python
messages = []
```

```python
add_user_message(messages, "What's 1+1?")

# Pass the list of messages into chat to get an answer
answer = chat(messages)
print(answer)

# Take the answer and add it as an assistant message into our list
add_assistant_message(messages, answer)

# Add in the user's followup question
add_user_message(messages, "And 3 more added to that?")

# Call chat again with the list of messages to get a final answer
answer = chat(messages)
print(answer)
```

## "Bot" Example

```python
messages = []

while True:
    user_input = input("> ")
    print("user input: ", user_input)
    add_user_message(messages, user_input)
    answer = chat(messages)
    add_assistant_message(messages, answer)
    print(answer)
```
