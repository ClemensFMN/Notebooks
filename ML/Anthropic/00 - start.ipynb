## Model Selection

I took the modelId from a bedrock tutorial & it seems that this is the only one working... In addition, I have to use the `us-east-1` region... Other regions do not work. I have no idea why => TODO

```python
client = boto3.client("bedrock", region_name="us-east-1")  
```

```python
models = client.list_foundation_models()
models
```

```python
modelId = "us.anthropic.claude-haiku-4-5-20251001-v1:0"

# modelId = "eu.anthropic.claude-haiku-4-5-20251001-v1:0"
```

```python
user_message = { 
            "role": "user", 
            "content": [{"text": "Write a one-sentence bedtime story about a unicorn."}]
        } 
```

```python
client = boto3.client("bedrock-runtime", region_name="us-east-1")  
```

```python
response = client.converse(modelId=modelId, messages=[user_message])
print(response)
```

```python
print(response["output"]["message"]["content"][0]["text"])
```
