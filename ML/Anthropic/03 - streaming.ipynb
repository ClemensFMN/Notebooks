## Model Selection

I took the modelId from a bedrock tutorial & it seems that this is the only one working... In addition, I have to use the `us-east-1` region... Other regions do not work. I have no idea why => TODO

```python
model_id = "us.anthropic.claude-haiku-4-5-20251001-v1:0"
```

```python
client = boto3.client("bedrock-runtime", region_name="us-east-1")
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": [
            {"text": text}
        ]
    }
    messages.append(user_message)


def chat(messages, system=None):
    params = {"modelId": model_id, "messages": messages}
    
    if system:
        params["system"] = [{"text": system}]
    
    response = client.converse_stream(**params)
    text = ""
    for event in response["stream"]:
        if "contentBlockDelta" in event:
            chunk = event["contentBlockDelta"]["delta"]["text"]
            print(chunk, end="")
            text += chunk

    print("\n\nTotal Message:\n" + text)
```

## Simple Example

```python
messages = []
```

```python
add_user_message(messages, "how can i host a postgres db?")

# Pass the list of messages into chat to get an answer
answer = chat(messages)
print(answer)
```
