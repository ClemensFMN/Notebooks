```python
from sklearn.datasets import make_blobs, make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
```

```python
n_samples = 100
centers = 4
```

```python
# X, y = make_classification(n_samples=n_samples, n_classes = 2, n_features=2, n_redundant=0, class_sep=1)

# random centers
# X, y = make_blobs(n_samples=n_samples, n_features=2, centers = centers)

# defined centers
X, y = make_blobs(n_samples=n_samples, n_features=2, centers = [[0,0], [1,5], [-2,-10]])
```

```python
for k in range(centers):
    members = y == k
    plt.plot(X[members, 0], X[members, 1], 'o')
plt.grid(True)
```

```python
clf = LogisticRegression(random_state=0).fit(X, y)
```

```python
DecisionBoundaryDisplay.from_estimator(clf, X, response_method="predict", plot_method="pcolormesh", shading="auto",)
for k in range(centers):
    members = y == k
    plt.plot(X[members, 0], X[members, 1], 'o')
```
