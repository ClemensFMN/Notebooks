```python
from sklearn.datasets import make_blobs, make_classification, make_moons
from sklearn.neural_network import MLPClassifier
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
```

```python
n_samples = 10000
centers = 4
```

```python
# defined centers
# X, y = make_blobs(n_samples=n_samples, n_features=2, centers = [[0,0], [1,5], [-2,-10]])

X, y = make_moons(n_samples=n_samples, noise=0.2, random_state=42)
```

```python
for k in range(centers):
    members = y == k
    plt.plot(X[members, 0], X[members, 1], 'o')
plt.grid(True)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10, 2), random_state=1)
clf.fit(X, y)
```

```python
DecisionBoundaryDisplay.from_estimator(clf, X, response_method="predict", plot_method="pcolormesh", shading="auto",)
for k in range(centers):
    members = y == k
    plt.plot(X[members, 0], X[members, 1], 'o')
```
