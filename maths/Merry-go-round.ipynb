```python
import numpy as np
import matplotlib.pyplot as plt
```

# Merry-go-round

We have a circle at the origin with radius $R$ which rotates with speed $\omega_R$. At position $(R,0)$, sits a smaller circle (radius $r$) which rotates with speed $\omega_r$. We want the $x-y$ coordinates of a fixed point on this second circle.

```python
R = 1
r = 0.1
wR = 1
wr = 21 # things change when the becomes eg 21.1!
```

```python
t = np.linspace(start = 0, stop = 50, num = 1000)
```

The following are the $x-y$ coordinates of the origin of the second (smaller) circle.

```python
Mx = R*np.cos(wR * t)
My = R*np.sin(wR * t)
```

```python
plt.plot(Mx, My)
```

Now we superposition the smaller circle onto the larger one.

```python
Px = R*np.cos(wR * t) + r*np.cos(wr * t)
Py = R*np.sin(wR * t) + r*np.sin(wr * t)
```

```python
plt.plot(Px, Py, '-x')
```
