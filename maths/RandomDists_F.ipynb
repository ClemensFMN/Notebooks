```python
import numpy as np
import scipy as sc
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Quotient of 2 Chi-squared RVs

U is distributed as \chi^2 with DoF df_u, V is distributed as \chi^2 with DoF df_v and we consider the quotient according to

$$
W = \frac{U / df_u}{V / df_v}
$$

$W$ is distributed with an f-distribution, $F_{df_u,df_v}$.

```python
# chi-squared dist
# df_v >> df_u takes rather long to execute!
df_u = 3
df_v = 6

U = stats.chi2(df_u)
V = stats.chi2(df_v)
```

```python
t = np.linspace(0, 10, 1000)
```

```python
plt.plot(t, U.pdf(t), '-r')
plt.plot(t, V.pdf(t), '-g')

# plt.axis([0, 10, 0, 0.8])
plt.grid(True)
```

```python
W = stats.f(df_u, df_v)
plt.plot(t, W.pdf(t), '-r')

# plt.axis([0, 10, 0, 0.8])
plt.grid(True)
```

```python
N = 1_000_000

us = U.rvs(N)
vs = V.rvs(N)

ws = (us / df_u) / (vs / df_v)

plt.hist(ws, density=True, bins='auto', histtype='stepfilled')


plt.plot(t, W.pdf(t), '-r')

plt.axis([0, 10, 0, 0.8])
plt.grid(True)
```
