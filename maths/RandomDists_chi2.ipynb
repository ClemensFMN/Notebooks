```python
import numpy as np
import scipy as sc
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Normal Distribution

```python
# generate a normal distribution (0,1)
n = stats.norm()
# and obtain RVs
x = n.rvs(1_000_000)
```

```python
# plot the histogram
plt.hist(x, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-5,5)
plt.plot(t, n.pdf(t))

plt.grid(True)
```

# Chi-Squared Distribution, Dof = 1

```python
# chi-squared dist
rv = stats.chi2(df = 1)
t = np.linspace(0, 4, 100)
plt.plot(t, rv.pdf(t), '-r')

# this should have a chi-sqaure distribution with DoF = 1
u = x**2
plt.hist(u, density=True, bins='auto', histtype='stepfilled')

plt.axis([0, 4, 0, 2])
plt.grid(True)
plt.savefig('2025-10-09-plt_2.png')
```

# Chi-Squred Distribution, Dof > 1

```python
# generate a normal distribution (0,1)
n = stats.norm()
# and obtain RVs
df = 5
N = 1_000_000
x = n.rvs((df, N))
# calculate the sum of df squared normal RVs
y = np.sum(x**2, axis=0)
```

```python
# chi-squared dist
rv = stats.chi2(df)
t = np.linspace(0, 25, 100)
plt.plot(t, rv.pdf(t), '-r')

plt.hist(y, density=True, bins='auto', histtype='stepfilled')

plt.grid(True)
```

```python
# chi-squared dist
df = 10
rv = stats.chi2(df)
rv.stats()
```

```python
t = np.linspace(0, 2*df, 100)
plt.plot(t, rv.pdf(t), '-r')
plt.grid(True)
```
