```python
import numpy as np
import scipy as sc
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
N = 1_000_000
```

```python
# generate a uniform distribution (0,1)
n = stats.uniform()
# and obtain RVs
x = n.rvs(N)
```

```python
# plot the histogram
plt.hist(x, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-1,2)
plt.plot(t, n.pdf(t))

plt.grid(True)
```

```python
y = x**2
```

```python
# plot the histogram
plt.hist(y, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(0,1)
plt.plot(t, 1/(2*np.sqrt(t)))

plt.grid(True)
```

```python
# generate a uniform distribution (-1,1)
n = stats.uniform(-1,2)
# and obtain RVs
x = n.rvs(N)
```

```python
# plot the histogram
plt.hist(x, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-2,2)
plt.plot(t, n.pdf(t))

plt.grid(True)
```

```python
y = x**2
```

```python
# plot the histogram
plt.hist(y, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(0,1)
plt.plot(t, 1/(2*np.sqrt(t)))

plt.grid(True)
plt.savefig('2025-10-09-plt_1.png')
```

```python
e = 0.02

print("analytical result = ", np.sqrt(e), "measured = ", len(y[y < e]) / N)
```
