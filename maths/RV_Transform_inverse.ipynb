```python
import numpy as np
import scipy as sc
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
N = 1_000_000
```

```python
# generate a uniform distribution (0,1)
n = stats.uniform()
# and obtain RVs
x = n.rvs(N)
```

```python
# plot the histogram
plt.hist(x, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-1,2)
plt.plot(t, n.pdf(t))

plt.grid(True)
```

```python
y = 1 / x
```

```python
# plot the histogram
plt.hist(y, density=True, bins=40, range=[0, 10], histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(1,10)
plt.plot(t, 1/ t**2)

plt.grid(True)
plt.savefig('2025-10-09-plt_3.png')
```

```python
# generate a uniform distribution (-1,1)
n = stats.uniform(-1,2)
# and obtain RVs
x = n.rvs(N)
```

```python
# plot the histogram
plt.hist(x, density=True, bins=40, histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-2,2)
plt.plot(t, n.pdf(t))

plt.grid(True)
```

```python
y = 1 / x
```

```python
# plot the histogram
plt.hist(y, density=True, bins=40, range=[-10,10], histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(1.3,10)
plt.plot(t, 1/(2*t**2), 'r')

t = np.linspace(-1.3,-10)
plt.plot(t, 1/(2*t**2), 'r')


plt.grid(True)
plt.savefig('2025-10-09-plt_4.png')
```

```python
# generate a normal distribution
n = stats.norm()
# and obtain RVs
x = n.rvs(N)
```

```python
y = 1/x
```

```python
# plot the histogram
plt.hist(y, density=True, bins=50, range=[-10,10], histtype='stepfilled')
# and compare it with the pdf
t = np.linspace(-10,10, 1000)
plt.plot(t, 1/( np.sqrt(2*np.pi) * t**2)*np.exp(-1/(2*t**2)))

plt.grid(True)
plt.savefig('2025-10-09-plt_5.png')
```
