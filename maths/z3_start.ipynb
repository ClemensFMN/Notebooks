# Fun with Z3 SOlver

```python
from z3 import *
```

```python
x = Int('x')
y = Int('y')
z = Int('z')
```

```python
# we solve the given model
# once we have found a solution, we add it as "negative" constraint, solve again & get the next solution
# inspired by https://gist.github.com/reb311ion/fefd07b933ac5376d5b531ada4fcc019
# I have no idea if this finds "all" solutions in a more complex setting
def getAllModels(s):
    res = s.check()
    if res == sat:
        print(s.model())
        s.add(Or([sym() != s.model()[sym] for sym in s.model().decls()]))
        # uncomment the following to see how we add the current solution as constraint
        # print(s)
        getAllModels(s)
    else:
        print("done")
```

## Example with 2 Variables

```python
# we get a new solver
s = Solver()
print(s)
```

```python
# add constraints & expressions to it
s.add(x < 10, y < 10, x + y == 10)
```

```python
print(s)
```

```python
# if the result of check == sat, z3 has found a solution
s.check()
```

```python
# which we can access / print
s.model()
```

```python
getAllModels(s)
```

## 3 Variables

```python
s = Solver()
s.add(x < 5, y < 2, z < 10, x + y + z == 10)
```

```python
getAllModels(s)
```
