```python
import numpy as np
import scipy as sc
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Quotient of normal RVs

```python
N = 10_000_000
```

```python
# generate a mormal distribution (0,1)
n = stats.norm()
# and obtain RVs
x = n.rvs(N)
y = n.rvs(N)
```

```python
z = x / y
```

```python
# plot the histogram
max_plot = 10
plt.hist(z, density=True, bins=50, range=[-max_plot,max_plot], histtype='stepfilled')
# and compare it with the pdf
zcalc = stats.cauchy()
t = np.linspace(-max_plot,max_plot)
plt.plot(t, zcalc.pdf(t))

plt.grid(True)
```

# Quotient of uniform RVs

```python
# generate a uniform distribution (0,1)
n = stats.uniform()
# and obtain RVs
x = n.rvs(N)
y = n.rvs(N)
```

```python
z = x / y
```

```python
# plot the histogram
max_plot = 5
plt.hist(z, density=True, bins=50, range=[0,max_plot], histtype='stepfilled')

# and compare it with the pdf
t = np.linspace(1,max_plot)
zcalc = 1/(2*t**2)
plt.plot(t, zcalc)

plt.grid(True)
```
