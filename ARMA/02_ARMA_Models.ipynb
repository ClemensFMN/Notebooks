# ARMA Models

```python
import numpy as np
from numpy.random import default_rng
import scipy.signal as sig
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools as stools
plt.rcParams["figure.figsize"] = (8,5)
rng = default_rng()
```

Some random input signal

```python
u = rng.standard_normal(30)
```

## MA Model

$$
y[n] = \sum_{k=0}^q b[k] u[n-k]
$$

```python
a = [1]
b = [1, 0.6, 0.2]

y = sig.lfilter(b,a,u)
plt.stem(y)
```

## AR Model

$$
y[n] = - \sum_{k=1}^p a[k] y[n-k]
$$

```python
a = [1, -0.8, 0.5]
b = [1]

y = sig.lfilter(b,a,u)
plt.stem(y)
```

Not much difference to be seen between the AR & MA model...

## Autocorrelation Estimation

```python
u = rng.standard_normal(10000)
a = [1]
b = [1, 0.6, 0.2]

yma = sig.lfilter(b,a,u)

a = [1, -0.8, 0.5]
b = [1]

yar = sig.lfilter(b,a,u)


acf_ma=stools.acf(yma, nlags=10, fft=False)
acf_ar=stools.acf(yar, nlags=10, fft=False)

plt.plot(acf_ma, '-rx', acf_ar, '-go')
plt.grid()
```
