# Lorentz Equation

Consider the ODE system

$$
\begin{align*}
x' &= \sigma (y-x) \\
y' &= rx - y - xz \\
z' &= xy - bz
\end{align*}
$$


```python
sigma = 10.
# chaos
r = 28.
# transient chaos
# r = 21.
b = 8/3
```

```python
def dX_dt(X, t=0):
    return np.array([ sigma * (X[1] - X[0]),
                           r * X[0] - X[1] - X[0]*X[2],
                           X[0] * X[1] - b*X[2]])    
```

```python
tmax = 70
t = np.linspace(0, tmax, 10000)
X, infodict = solveODE(dX_dt, [0.0, 1.0, 0.0], t)
infodict['message']
```

```python
x, y, z = X.T

plt.plot(t, x, 'g-', label='$x(t)$')
plt.plot(t, y, 'r-', label='$y(t)$')
plt.plot(t, z, 'b-', label='$z(t)$')

plt.grid()
plt.legend(loc='best')
plt.xlabel('$t$')
plt.ylabel('$y$')
```
