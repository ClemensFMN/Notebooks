## ODE System

$$
\begin{align*}
x' &= y \\
y' &= x^3 - x\\
\end{align*}
$$

Fixed points are located at $(0,0), (\pm 1, 0)$

```python
def dX_dt(X, t=0):
    return np.array([X[1],
                     X[0]**3 - X[0]])
```

```python
vField(-2, 2, -2, 2, dX_dt, Norm = True)
```

## Jacboi Matrix

$$
J = \begin{pmatrix} 0 & 1 \\ 3x^2-1 & 0\end{pmatrix}
$$

J at $(0,0)$:

$$
J = \begin{pmatrix} 0 & 1 \\ -1 & 0\end{pmatrix}
$$

J at $(\pm 1,0)$

$$
J = \begin{pmatrix} 0 & 1 \\ 2 & 0\end{pmatrix}
$$


```python
# (0,0)
J = np.array([[0, 1],[-1,0]])
lin.eig(J)
```

```python
# (\pm 1,0)
J = np.array([[0, 1],[2,0]])
lin.eig(J)
```
