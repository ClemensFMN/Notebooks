```python
def dX_dt(X, t=0):
    return np.array([-2*np.cos(X[0]) - np.cos(X[1]) ,
                     -2*np.cos(X[1]) - np.cos(X[0])  ])
```

```python
vField(-3, 3, -3, 3, dX_dt, Norm = True)
```

Fixed points are located at

$$
(x,y) = (\pm \pi/2, \pm \pi/2)
$$

The Jacobian matrix has the following form

$$
A = \begin{pmatrix} 2 \sin(x) & \sin(y) \\ \sin(x) & 2 \sin(y) \end{pmatrix}
$$

At the fixed point $(- \pi/2, - \pi/2)$ it has the following value

$$
A = \begin{pmatrix} 2 \sin(- \pi/2) & \sin(- \pi/2) \\ \sin(- \pi/2) & 2 \sin(- \pi/2) \end{pmatrix} = \begin{pmatrix} -2 & -1 \\ -1  & - 2 \end{pmatrix}
$$

Let's calculate the eigenvalues.

```python
A = np.array([[-2., -1.], [-1., -2.]])
lin.eig(A)
```

They are both negative, so the fixed point is a stable one.

Let's check the other fixed points.

```python
x = np.pi/2
y = np.pi/2

A = np.array([[2*np.sin(x) , np.sin(y)], [np.sin(x), 2*np.sin(y)]])
lin.eig(A)
```

```python
x = np.pi/2
y = -np.pi/2

A = np.array([[2*np.sin(x) , np.sin(y)], [np.sin(x), 2*np.sin(y)]])
lin.eig(A)
```

```python
x = -np.pi/2
y = np.pi/2

A = np.array([[2*np.sin(x) , np.sin(y)], [np.sin(x), 2*np.sin(y)]])
lin.eig(A)
```
