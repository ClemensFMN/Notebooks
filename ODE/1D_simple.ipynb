```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
import scipy.integrate as int
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (7,5)
```

Consider

$$
x' = (x-2)*(x+1)
$$

which has two fixed points, $x = -1, x = 2$. Stability of these is determined by

$$
x'' = 2x-1
$$

At $x = -1$, we have $x'' = -3 < 0$, therefore the fixed point is stable. At the other, $x=2$, we have $x'' = 3 > 0$ and therefore the point is instable.

```python
def dX_dt(X, t=0):
    return np.array([ (X[0]-2)*(X[0] + 1)])
```

```python
t  = np.linspace(0, 10, 1000)
y0 = [1.]

X, infodict = int.odeint(dX_dt, y0, t, full_output=True)
infodict['message']
```

```python
y = X

plt.plot(t, y  , 'b-', label='y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('y')
```
