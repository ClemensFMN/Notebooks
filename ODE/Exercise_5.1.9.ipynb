We consider the linear ODE system

\begin{align*}
x' &= -y \\
y' &= -x
\end{align*}

```python
def dX_dt(X, t=0):
    return np.array([       -X[1] ,
                     -X[0]         ])
```

```python
t  = np.linspace(0, 20, 1000)

X, infodict = solveODE(dX_dt, [1., 1.], t)
infodict['message']
```

```python
y1, y2 = X.T

plt.plot(t, y1, 'r-', label='$y_1$')
plt.plot(t, y2  , 'b-', label='$y_2$')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('$y_1, y_2$')
```

The point $(0,0)$ is a saddle point (and the only fixed point).

The trajectories towards this point are lines and therefore have the form

$$
\mathbf{x} = C \mathbf{v} e^{\lambda t}
$$

where $\mathbf{v}$ is an eigenvector and $\lambda$ is the corresponding eigenvalue.

```python
A = np.array([[0., -1.], [-1., 0.]])
```

```python
lin.eig(A)
```

The eigenvalue $1$ is associated with the eigenvector $(1, -1)^T$ and the corresponding solutions are therefore

$$
\mathbf{x} = C(1, -1)^T e^{t}
$$

The other eigenvalue $-1$ is associated with the eigenvector $(1, 1)^T$ and the corresponding solutions are therefore

$$
\mathbf{x} = C(1, 1)^T e^{-t}
$$

In the book's example, they propose to introduce new variables $u = x+y, v = x-y$ which decouples the system. This corresponds a coordinate transformation in direction of the eigenvectors and therefore simplifies the ODE system / solutions.
