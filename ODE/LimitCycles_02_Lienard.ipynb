Consider the Lienard system

\begin{align*}
x' &= y \\
y' &= -g(x) - f(x)y
\end{align*}

Setting  $f(x) = - \mu(1-x^2)$ and $g(x) = x$ yields the van der Pol oscillator which we have covered in Exercise_6.1.8 (van der Pol). There we have also played around with different damping terms $f(x)$.

Here we want to consider different functions for $g(x)$.

```python
mu = 2.0
```

```python
# van der Pol
def f(x):
    return(-mu*(1-x**2))

def g(x):
    return(x)
```

```python
# different g(x)
def f(x):
    return(-mu*(1-x**2))

def g(x): # has to be an odd function for the ODE system to have a limit cycle
    return(x**3)
```

```python
def dX_dt(X, t=0):
    return np.array([X[1],
                     -g(X[0]) - f(X[0]) * X[1] ])
```

```python
vField(-2, 2, -5, 5, dX_dt, Norm = False)
```
