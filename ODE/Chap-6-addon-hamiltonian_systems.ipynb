# Hamiltonian Systems

We have a Hamiltonian function $H(x,y)$ and obtain from this a 2D ODE system,

$$
\begin{align*}
x' &= \frac{\partial H}{\partial y} \\
y' &= -\frac{\partial H}{\partial x}
\end{align*}
$$

As a simple example we consider the undamped oscillator, 

$$
\begin{align*}
x' &= y \\
y' &= -kx
\end{align*}
$$

which is a Hamiltonian system with $H(x,y) = \frac{1}{2} y^2 + \frac{k}{2} x^2$.


```python
k = 1
```

```python
def dX_dt(X, t=0):
    return np.array([X[1],
                     -k*X[0]  ])
```

```python
mx = 5
vField(-mx, mx, -mx, mx, dX_dt, Norm = True)
```

```python
tmax = 7
t = np.linspace(0, tmax, 200)
X, infodict = solveODE(dX_dt, [1.99, 0.99], t)
infodict['message']
```

```python
x, y = X.T

plt.plot(t, x, 'r-', label='$x$')
plt.plot(t, y  , 'b-', label='$y$')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('$x, y$')
```

```python
plt.plot(x,y)
plt.grid(True)
```

The important thing about Hamiltonian systems is that $H(x,y)$ is constant along every solution. We can see this from

$$
\frac{dH}{dt} = \frac{\partial H}{\partial x} \frac{dx}{dt} + \frac{\partial H}{\partial y} \frac{dy}{dt}
$$

Inserting the definition of $x'$ and $y'$ yields

$$
\frac{dH}{dt} = \frac{\partial H}{\partial x} \frac{\partial H}{\partial y} - \frac{\partial H}{\partial y} \frac{\partial H}{\partial x} = 0
$$


```python
def H(x,y):
    return(0.5*y**2 + k/2*x**2)
```

```python
plt.plot(H(x,y))
plt.ylim([0,3])
plt.grid(True)
```
