# Predator-Pre System with finite capacity

Consider the ODE system

$$
\begin{align*}
x' &= x(a - by - \lambda x) \\
y' &= y(-c + dx - \mu y)
\end{align*}
$$

Let's consider the nullclines:

* x-axis
* y-axis
* L: $a - by - \lambda x = 0$
* M: $-c + dx - \mu y = 0$

```python
# CASE 1: L and M have an interseciton in x, y > 0
a = 10
b = 2
lmbda = 1

c = 20
d = 3
mu = 0.5
```

```python
# CASE 2: L and M have NO interseciton in x, y > 0
a = 5
b = 2
lmbda = 1

c = 20
d = 3
mu = 0.5
```

```python
def dX_dt(X, t=0):
    return np.array([X[0] * (a -  b*X[1] - lmbda*X[0]) ,
                     X[1] * (-c + d*X[0] - mu *   X[1])])
```

```python
vField(0, 10, 0, 10, dX_dt, Norm = True)

x = np.linspace(0,10,100)
yn1 = 1/b*(a - lmbda*x)
yn2 = 1/mu*(-c+d*x)
plt.plot(x,yn1)
plt.plot(x,yn2)
plt.xlim(0,10)
plt.ylim(0,10)
```

```python
tmax = 20
t = np.linspace(0, tmax, 200)

X, infodict = solveODE(dX_dt, [1.0, 1.0], t)


infodict['message']
```

```python
x, y = X.T

plt.plot(t, x, 'r-', label='$x(t)$')
plt.plot(t, y, 'g-', label='$y(t)$')

plt.grid()
plt.legend(loc='best')
plt.xlabel('$t$')
plt.ylabel('$x(t), y(t)$')
```
