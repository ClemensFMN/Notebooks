```python
sigma, r, b = 10., 28., 8/3
```

```python
def dX_dt(X, t=0):
    return np.array([sigma*(X[1] - X[0]),
                     r*X[0] - X[1] - X[0]*X[2],
                     X[0]*X[1] - b*X[2] ])
```

```python
tmax = 70
t = np.linspace(0, tmax, 5000)
X, infodict = solveODE(dX_dt, [0., 1.0, 0.], t)
infodict['message']
```

```python
x, y, z = X.T
```

```python
plt.plot(t, x, 'r-', label='$x$')
plt.plot(t, y, 'g-', label='$y$')
plt.plot(t, z, 'b-', label='$z$')

plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
```
