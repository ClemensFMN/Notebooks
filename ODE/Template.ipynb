```python
def dX_dt(X, t=0):
    return np.array([X[1] + X[1]**2,
                     -1/2*X[0] + 1/5*X[1] - X[0]*X[1] + 6/5*X[1]**2 ])
```

```python
mx = 5
vField(-mx, mx, -mx, mx, dX_dt, Norm = True)
```

```python
tmax = 7
t = np.linspace(0, tmax, 200)
X, infodict = solveODE(dX_dt, [1.99, 0.99], t)
infodict['message']
```

```python
x, y = X.T

plt.plot(t, x, 'r-', label='$x$')
plt.plot(t, y  , 'b-', label='$y$')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('$x, y$')
plt.ylim([0,5])
```
